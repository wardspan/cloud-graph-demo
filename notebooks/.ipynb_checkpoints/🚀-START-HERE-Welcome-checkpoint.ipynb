## ğŸ”— **Test Your Connection**

Let's verify your setup is working correctly:

````python
# Test connection to the Cloud Threat Graph Lab database
try:
    from neo4j import GraphDatabase
    import pandas as pd
    
    # Connect to Neo4j
    driver = GraphDatabase.driver("bolt://neo4j:7687", auth=("neo4j", "cloudsecurity"))
    
    with driver.session() as session:
        result = session.run("""
        MATCH (n) 
        RETURN labels(n)[0] as NodeType, count(n) as Count 
        ORDER BY Count DESC 
        LIMIT 5
        """)
        
        data = [record.data() for record in result]
        df = pd.DataFrame(data)
        
    driver.close()
    
    print("ğŸ‰ SUCCESS! Connected to Cloud Threat Graph Lab Database")
    print("\nğŸ“Š Security Asset Inventory (Top 5):")
    print(df.to_string(index=False))
    print("\nâœ… Your environment is ready for security analysis!")
    
except Exception as e:
    print(f"âš ï¸  Connection issue: {e}")
    print("\nğŸ”§ Troubleshooting:")
    print("   1. Make sure Docker services are running")
    print("   2. Wait a few minutes for services to start")
    print("   3. Try restarting: docker-compose restart")
````
